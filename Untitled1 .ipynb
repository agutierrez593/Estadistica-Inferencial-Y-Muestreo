```python
import pandas as pd

# Cargar archivo CSV con codificación 'latin-1' primero
try:
    df = pd.read_csv("/content/nacidos.csv", encoding='latin-1')
except UnicodeDecodeError:
    # If latin-1 fails, try utf-8 as a fallback
    df = pd.read_csv("/content/nacidos.csv", encoding='utf-8')


# Identificar columnas numéricas para la detección de outliers
numerical_cols = ["PESO (Gramos)", "TALLA (CentÃ­metros)", "NUMERO DE HIJOS NACIDOS VIVOS"]

# Crear una copia del DataFrame original para no modificarlo directamente
df_cleaned = df.copy()

# Eliminar outliers usando el método IQR para cada columna numérica
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar el DataFrame, manteniendo solo los valores dentro de los límites
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

# Mostrar el número de filas antes y después de eliminar outliers
print(f"Número de filas antes de eliminar outliers: {df.shape[0]}")
print(f"Número de filas después de eliminar outliers: {df_cleaned.shape[0]}")

# Mostrar información básica y resumen estadístico del nuevo DataFrame limpio
print("\nInformación del DataFrame después de eliminar outliers:")
print(df_cleaned.info())
print("\nResumen estadístico del DataFrame después de eliminar outliers:")
print(df_cleaned.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualización de variables numéricas del DataFrame limpio
numerical_cols = ["PESO (Gramos)", "TALLA (CentÃ­metros)", "NUMERO DE HIJOS NACIDOS VIVOS"]
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df_cleaned, x=col, kde=True)
    plt.title(f'Histograma de {col} (Datos sin outliers)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df_cleaned, y=col)
    plt.title(f'Diagrama de Caja de {col} (Datos sin outliers)')
    plt.ylabel(col)
    plt.show()
```

```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm

# Cargar archivo CSV con codificación 'latin-1' primero
try:
    df = pd.read_csv("/content/nacidos.csv", encoding='latin-1')
except UnicodeDecodeError:
    # If latin-1 fails, try utf-8 as a fallback
    df = pd.read_csv("/content/nacidos.csv", encoding='utf-8')

# Identificar columnas numéricas para la detección de outliers
# Using the corrected column name based on previous outputs
numerical_cols_outliers = ["PESO (Gramos)", "TALLA (CentÃ­metros)", "NUMERO DE HIJOS NACIDOS VIVOS"]

# Crear una copia del DataFrame original para no modificarlo directamente
df_cleaned = df.copy()

# Eliminar outliers usando el método IQR para cada columna numérica
for col in numerical_cols_outliers:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar el DataFrame, manteniendo solo los valores dentro de los límites
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]


# Identificar columnas numéricas continuas para los tests de normalidad
# Excluyendo 'NUMERO DE HIJOS NACIDOS VIVOS' ya que es discreta
numerical_continuous_cols = ["PESO (Gramos)", "TALLA (CentÃ­metros)"]

print("Resultados de los Tests de Normalidad:")

for col in numerical_continuous_cols:
    print(f"\n--- Variable: {col} ---")

    # 1) Test de Shapiro-Wilk
    shapiro_test = stats.shapiro(df_cleaned[col])
    print(f"Shapiro-Wilk: Estadístico={shapiro_test.statistic:.4f}, p-valor={shapiro_test.pvalue:.4f}")

    # 2) Test de Kolmogorov-Smirnov (usando Lilliefors para la normalidad)
    # stats.kstest requiere especificar la distribución, y Lilliefors es para normalidad cuando media y varianza son desconocidas
    # Una forma común es usar statsmodels.api.lilliefors
    try:
        lilliefors_test = sm.stats.lilliefors(df_cleaned[col])
        print(f"Kolmogorov-Smirnov (Lilliefors): Estadístico={lilliefors_test[0]:.4f}, p-valor={lilliefors_test[1]:.4f}")
    except Exception as e:
        print(f"No se pudo realizar el test de Lilliefors para {col}: {e}")

    # 3) Test de Anderson-Darling
    anderson_test = stats.anderson(df_cleaned[col], dist='norm')
    print(f"Anderson-Darling: Estadístico={anderson_test.statistic:.4f}")
    for i in range(len(anderson_test.critical_values)):
        sl, cv = anderson_test.significance_level[i], anderson_test.critical_values[i]
        print(f"  Nivel de significancia {sl}%: Valor crítico={cv:.4f}")

    # 4) Test de Jarque-Bera
    # stats.jarque_bera requiere que los datos no tengan valores nan
    jarque_bera_test = stats.jarque_bera(df_cleaned[col].dropna())
    print(f"Jarque-Bera: Estadístico={jarque_bera_test.statistic:.4f}, p-valor={jarque_bera_test.pvalue:.4f}")
```
