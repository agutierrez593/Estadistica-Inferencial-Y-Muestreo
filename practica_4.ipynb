1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.


```python
import numpy as np
from scipy.stats import norm

# Datos
media_muestra = 170  # cm
desviacion_estandar_poblacional = 10  # cm
tamano_muestra = 50
nivel_confianza = 0.95

# Nivel de significancia
alpha = 1 - nivel_confianza

# Valor Z crítico para un intervalo de confianza bilateral
# Usamos la función ppf (percent point function) que es la inversa de la cdf
valor_z_critico = norm.ppf(1 - alpha/2)

# Margen de error
margen_error = valor_z_critico * (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

# Intervalo de confianza
intervalo_confianza_inferior = media_muestra - margen_error
intervalo_confianza_superior = media_muestra + margen_error

print(f"La media de la muestra es: {media_muestra} cm")
print(f"La desviación estándar poblacional es: {desviacion_estandar_poblacional} cm")
print(f"El tamaño de la muestra es: {tamano_muestra}")
print(f"El nivel de confianza es: {nivel_confianza}")
print(f"El valor Z crítico es: {valor_z_critico:.4f}")
print(f"El margen de error es: {margen_error:.4f} cm")
print(f"\nEl intervalo de confianza del {nivel_confianza*100}% para la altura promedio poblacional es: ({intervalo_confianza_inferior:.4f}, {intervalo_confianza_superior:.4f}) cm")
```

11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72 % de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3 % y un
nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?


```python
import numpy as np
from scipy.stats import norm

# Datos del problema 11

# Con estudio piloto
proporcion_muestra = 0.72
margen_error = 0.03
nivel_confianza = 0.95

# Valor Z crítico para un intervalo de confianza bilateral
alpha = 1 - nivel_confianza
valor_z_critico = norm.ppf(1 - alpha/2)

# Tamaño de muestra necesario con estudio piloto
# Fórmula: n = (Z^2 * p * (1-p)) / E^2
tamano_muestra_piloto = (valor_z_critico**2 * proporcion_muestra * (1 - proporcion_muestra)) / margen_error**2

# Redondear hacia arriba para asegurar el margen de error
tamano_muestra_piloto = np.ceil(tamano_muestra_piloto)

print(f"Con estudio piloto (proporción estimada = {proporcion_muestra}):")
print(f"El valor Z crítico es: {valor_z_critico:.4f}")
print(f"El margen de error deseado es: {margen_error}")
print(f"El tamaño de muestra necesario es: {int(tamano_muestra_piloto)}")

print("-" * 30)

# Sin estudio piloto
# Se asume la proporción más conservadora p = 0.5 para maximizar el tamaño de muestra
proporcion_conservadora = 0.5

# Tamaño de muestra necesario sin estudio piloto
# Fórmula: n = (Z^2 * 0.25) / E^2  (ya que p*(1-p) es máximo cuando p=0.5, resultando en 0.25)
tamano_muestra_sin_piloto = (valor_z_critico**2 * proporcion_conservadora * (1 - proporcion_conservadora)) / margen_error**2
tamano_muestra_sin_piloto = np.ceil(tamano_muestra_sin_piloto)

print(f"Sin estudio piloto (proporción conservadora = {proporcion_conservadora}):")
print(f"El valor Z crítico es: {valor_z_critico:.4f}")
print(f"El margen de error deseado es: {margen_error}")
print(f"El tamaño de muestra necesario es: {int(tamano_muestra_sin_piloto)}")
```

21. Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que
la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una
media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia
α = 0,05.


```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Datos del problema 21
media_poblacional_hipotetizada = 500  # g (afirmación de la empresa)
media_muestra = 495  # g (media de la muestra)
desviacion_estandar_poblacional = 15  # g
tamano_muestra = 40
nivel_significancia_alpha = 0.05

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La media poblacional es igual a 500 g (la máquina está calibrada).
# Hipótesis Alternativa (H1): La media poblacional es menor que 500 g (la máquina está subllenando).
# Esta es una prueba unilateral a la izquierda.

print("1. Hipótesis:")
print(f"   H0: μ = {media_poblacional_hipotetizada} g")
print(f"   H1: μ < {media_poblacional_hipotetizada} g (Prueba unilateral a la izquierda)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es n >= 30,
# por lo tanto, usamos el estadístico Z y la distribución normal estándar.
print("\n3. Estadístico de prueba: Z (debido a que se conoce σ y n >= 30)")
print("   Distribución: Normal estándar")

# 4. Valor crítico y región de rechazo
# Para una prueba unilateral a la izquierda con α = 0.05, buscamos el valor Z tal que el área a su izquierda sea 0.05.
valor_z_critico = norm.ppf(nivel_significancia_alpha)

print(f"\n4. Valor crítico: Z_{nivel_significancia_alpha} = {valor_z_critico:.4f}")
print(f"   Región de rechazo: Z_calculado < {valor_z_critico:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Z para la media: Z = (media_muestra - media_poblacional_hipotetizada) / (σ / sqrt(n))
estadistico_z_calculado = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

# El valor p para una prueba unilateral a la izquierda es el área a la izquierda del estadístico Z calculado.
valor_p = norm.cdf(estadistico_z_calculado)

print(f"\n5. Estadístico de prueba calculado: Z = {estadistico_z_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar (Z)')

# Región de rechazo (área a la izquierda del valor crítico)
x_reject = np.linspace(-4, valor_z_critico, 100)
plt.fill_between(x_reject, norm.pdf(x_reject, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alpha})')

# Valor crítico
plt.axvline(valor_z_critico, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_z_critico:.4f}')

# Estadístico calculado
plt.axvline(estadistico_z_calculado, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z_calculado:.4f}')

# Marcar el valor p en la gráfica
plt.fill_between(np.linspace(-4, estadistico_z_calculado, 100), norm.pdf(np.linspace(-4, estadistico_z_calculado, 100), 0, 1), color='green', alpha=0.5, label=f'Valor p = {valor_p:.4f}')

plt.title('Prueba de Hipótesis Unilateral a la Izquierda (Z)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if valor_z_critico > estadistico_z_calculado:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = "Hay suficiente evidencia estadística, con un nivel de significancia del 5%, para concluir que la media poblacional de los paquetes de cereal es menor que 500 g. Esto sugiere que la máquina está descalibrada y subllenando."
else:
    decision = "No rechazar H0"
    conclusion = "No hay suficiente evidencia estadística, con un nivel de significancia del 5%, para concluir que la media poblacional de los paquetes de cereal es menor que 500 g. No hay evidencia para afirmar que la máquina está subllenada."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una media muestral de 495 g (o una más extrema) si la hipótesis nula (μ = 500 g) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es menor que el nivel de significancia (α = {nivel_significancia_alpha}), rechazamos la hipótesis nula.")
```

23. Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Datos del problema 23
media_poblacional_hipotetizada_p23 = 75  # puntos (promedio nacional)
media_muestra_p23 = 78  # puntos (media de la muestra)
desviacion_estandar_poblacional_p23 = 12  # puntos
tamano_muestra_p23 = 36
nivel_significancia_alpha_p23 = 0.05

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La media poblacional es igual a 75 puntos (el promedio de la clase es igual al nacional).
# Hipótesis Alternativa (H1): La media poblacional es diferente a 75 puntos (el promedio de la clase es diferente al nacional).
# Esta es una prueba bilateral (de dos colas).

print("1. Hipótesis:")
print(f"   H0: μ = {media_poblacional_hipotetizada_p23} puntos")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada_p23} puntos (Prueba bilateral)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha_p23}")

# 3. Estadístico de prueba y distribución
# Se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es n >= 30,
# por lo tanto, usamos el estadístico Z y la distribución normal estándar.
print("\n3. Estadístico de prueba: Z (debido a que se conoce σ y n >= 30)")
print("   Distribución: Normal estándar")

# 4. Valor crítico y región de rechazo
# Para una prueba bilateral con α = 0.05, el nivel de significancia se divide en dos colas (α/2 = 0.025).
# Buscamos los valores Z tales que el área a la izquierda de -Z_critico sea 0.025 y el área a la derecha de Z_critico sea 0.025.
valor_z_critico_p23 = norm.ppf(1 - nivel_significancia_alpha_p23/2)
valor_z_critico_negativo_p23 = -valor_z_critico_p23

print(f"\n4. Valores críticos: Z_{nivel_significancia_alpha_p23/2:.3f} = ±{valor_z_critico_p23:.4f}")
print(f"   Región de rechazo: Z_calculado < {valor_z_critico_negativo_p23:.4f} o Z_calculado > {valor_z_critico_p23:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Z para la media: Z = (media_muestra - media_poblacional_hipotetizada) / (σ / sqrt(n))
estadistico_z_calculado_p23 = (media_muestra_p23 - media_poblacional_hipotetizada_p23) / (desviacion_estandar_poblacional_p23 / np.sqrt(tamano_muestra_p23))

# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico Z calculado.
# Si Z_calculado es positivo, el valor p es 2 * (1 - cdf(Z_calculado)).
# Si Z_calculado es negativo, el valor p es 2 * cdf(Z_calculado)).
valor_p_p23 = 2 * (1 - norm.cdf(abs(estadistico_z_calculado_p23)))

print(f"\n5. Estadístico de prueba calculado: Z = {estadistico_z_calculado_p23:.4f}")
print(f"   Valor p: {valor_p_p23:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar (Z)')

# Región de rechazo (área en las colas)
x_reject_left = np.linspace(-4, valor_z_critico_negativo_p23, 100)
plt.fill_between(x_reject_left, norm.pdf(x_reject_left, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia_alpha_p23/2})')

x_reject_right = np.linspace(valor_z_critico_p23, 4, 100)
plt.fill_between(x_reject_right, norm.pdf(x_reject_right, 0, 1), color='red', alpha=0.5) # No repetir label

# Valores críticos
plt.axvline(valor_z_critico_negativo_p23, color='red', linestyle='--', label=f'Valores Críticos Z = ±{valor_z_critico_p23:.4f}')
plt.axvline(valor_z_critico_p23, color='red', linestyle='--') # No repetir label

# Estadístico calculado
plt.axvline(estadistico_z_calculado_p23, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z_calculado_p23:.4f}')

# Marcar el valor p en la gráfica (área en las colas más extremas que el estadístico calculado)
x_p_left = np.linspace(-4, -abs(estadistico_z_calculado_p23), 100)
plt.fill_between(x_p_left, norm.pdf(x_p_left, 0, 1), color='green', alpha=0.5, label=f'Valor p/2 = {valor_p_p23/2:.4f}')

x_p_right = np.linspace(abs(estadistico_z_calculado_p23), 4, 100)
plt.fill_between(x_p_right, norm.pdf(x_p_right, 0, 1), color='green', alpha=0.5) # No repetir label


plt.title('Prueba de Hipótesis Bilateral (Z)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if abs(estadistico_z_calculado_p23) > valor_z_critico_p23:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha_p23*100}%, para concluir que la media poblacional de la clase es diferente al promedio nacional de {media_poblacional_hipotetizada_p23} puntos."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha_p23*100}%, para concluir que la media poblacional de la clase es diferente al promedio nacional de {media_poblacional_hipotetizada_p23} puntos. No podemos afirmar que el promedio de la clase sea diferente."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p_p23:.4f}): El valor p es la probabilidad de obtener una media muestral de {media_muestra_p23} puntos (o una más extrema en cualquier dirección) si la hipótesis nula (μ = {media_poblacional_hipotetizada_p23} puntos) fuera verdadera. Dado que el valor p ({valor_p_p23:.4f}) es {'menor' if valor_p_p23 < nivel_significancia_alpha_p23 else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha_p23}), {'rechazamos' if valor_p_p23 < nivel_significancia_alpha_p23 else 'no rechazamos'} la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Datos del problema 25
media_poblacional_hipotetizada = 500  # mL (afirmación de la máquina)
media_muestra = 498  # mL (media de la muestra)
desviacion_estandar_poblacional = 5  # mL (σ)
tamano_muestra = 60
nivel_significancia_alpha = 0.01

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La media poblacional es igual a 500 mL (la máquina está llenando correctamente).
# Hipótesis Alternativa (H1): La media poblacional es diferente a 500 mL (la máquina no está llenando correctamente).
# Esta es una prueba bilateral (de dos colas).

print("1. Hipótesis:")
print(f"   H0: μ = {media_poblacional_hipotetizada} mL")
print(f"   H1: μ ≠ {media_poblacional_hipotetizada} mL (Prueba bilateral)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Se conoce la desviación estándar poblacional (σ) y el tamaño de la muestra es n >= 30,
# por lo tanto, usamos el estadístico Z y la distribución normal estándar.
print("\n3. Estadístico de prueba: Z (debido a que se conoce σ y n >= 30)")
print("   Distribución: Normal estándar")

# 4. Valor crítico y región de rechazo
# Para una prueba bilateral con α = 0.01, el nivel de significancia se divide en dos colas (α/2 = 0.005).
# Buscamos los valores Z tales que el área a la izquierda de -Z_critico sea 0.005 y el área a la derecha de Z_critico sea 0.005.
valor_z_critico = norm.ppf(1 - nivel_significancia_alpha/2)
valor_z_critico_negativo = -valor_z_critico

print(f"\n4. Valores críticos: Z_{nivel_significancia_alpha/2:.3f} = ±{valor_z_critico:.4f}")
print(f"   Región de rechazo: Z_calculado < {valor_z_critico_negativo:.4f} o Z_calculado > {valor_z_critico:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Z para la media: Z = (media_muestra - media_poblacional_hipotetizada) / (σ / sqrt(n))
estadistico_z_calculado = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico Z calculado.
# Si Z_calculado es positivo, el valor p es 2 * (1 - cdf(Z_calculado)).
# Si Z_calculado es negativo, el valor p es 2 * cdf(Z_calculado)).
valor_p = 2 * (1 - norm.cdf(abs(estadistico_z_calculado)))

print(f"\n5. Estadístico de prueba calculado: Z = {estadistico_z_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar (Z)')

# Región de rechazo (área en las colas)
x_reject_left = np.linspace(-4, valor_z_critico_negativo, 100)
plt.fill_between(x_reject_left, norm.pdf(x_reject_left, 0, 1), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia_alpha/2})')

x_reject_right = np.linspace(valor_z_critico, 4, 100)
plt.fill_between(x_reject_right, norm.pdf(x_reject_right, 0, 1), color='red', alpha=0.5) # No repetir label

# Valores críticos
plt.axvline(valor_z_critico_negativo, color='red', linestyle='--', label=f'Valores Críticos Z = ±{valor_z_critico:.4f}')
plt.axvline(valor_z_critico, color='red', linestyle='--') # No repetir label

# Estadístico calculado
plt.axvline(estadistico_z_calculado, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z_calculado:.4f}')

# Marcar el valor p en la gráfica (área en las colas más extremas que el estadístico calculado)
x_p_left = np.linspace(-4, -abs(estadistico_z_calculado), 100)
plt.fill_between(x_p_left, norm.pdf(x_p_left, 0, 1), color='green', alpha=0.5, label=f'Valor p/2 = {valor_p/2:.4f}')

x_p_right = np.linspace(abs(estadistico_z_calculado), 4, 100)
plt.fill_between(x_p_right, norm.pdf(x_p_right, 0, 1), color='green', alpha=0.5) # No repetir label


plt.title('Prueba de Hipótesis Bilateral (Z)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if abs(estadistico_z_calculado) > valor_z_critico:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la media poblacional del llenado de botellas es diferente a {media_poblacional_hipotetizada} mL. Esto sugiere que la máquina podría estar descalibrada."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la media poblacional del llenado de botellas es diferente a {media_poblacional_hipotetizada} mL. No hay suficiente evidencia para afirmar que la máquina está descalibrada."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una media muestral de {media_muestra} mL (o una más extrema en cualquier dirección) si la hipótesis nula (μ = {media_poblacional_hipotetizada} mL) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia_alpha else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha}), {'rechazamos' if valor_p < nivel_significancia_alpha else 'no rechazamos'} la hipótesis nula.")
```

estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.



```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Datos del problema 26
media_poblacional_hipotetizada = 25  # minutos (afirmación del restaurante)
media_muestra = 28  # minutos (media de la muestra)
desviacion_estandar_muestra = 4  # minutos (s)
tamano_muestra = 15
nivel_significancia_alpha = 0.05

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La media poblacional es igual a 25 minutos.
# Hipótesis Alternativa (H1): La media poblacional es mayor que 25 minutos (sospecha del cliente).
# Esta es una prueba unilateral a la derecha.

print("1. Hipótesis:")
print(f"   H0: μ = {media_poblacional_hipotetizada} minutos")
print(f"   H1: μ > {media_poblacional_hipotetizada} minutos (Prueba unilateral a la derecha)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Se desconoce la desviación estándar poblacional (σ) y el tamaño de la muestra es n < 30,
# por lo tanto, usamos el estadístico t y la distribución t de Student.
print("\n3. Estadístico de prueba: t (debido a que se desconoce σ y n < 30)")
print("   Distribución: t de Student")

# Grados de libertad
grados_libertad = tamano_muestra - 1
print(f"   Grados de libertad: {grados_libertad}")


# 4. Valor crítico y región de rechazo
# Para una prueba unilateral a la derecha con α = 0.05 y grados de libertad = 14,
# buscamos el valor t tal que el área a su derecha sea 0.05.
valor_t_critico = t.ppf(1 - nivel_significancia_alpha, grados_libertad)


print(f"\n4. Valor crítico: t_{nivel_significancia_alpha:.3f} ({grados_libertad} gl) = {valor_t_critico:.4f}")
print(f"   Región de rechazo: t_calculado > {valor_t_critico:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico t para la media: t = (media_muestra - media_poblacional_hipotetizada) / (s / sqrt(n))
estadistico_t_calculado = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_muestra / np.sqrt(tamano_muestra))

# El valor p para una prueba unilateral a la derecha es el área a la derecha del estadístico t calculado.
valor_p = 1 - t.cdf(estadistico_t_calculado, grados_libertad)


print(f"\n5. Estadístico de prueba calculado: t = {estadistico_t_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, t.pdf(x, grados_libertad), label=f'Distribución t de Student ({grados_libertad} gl)')

# Región de rechazo (área a la derecha del valor crítico)
x_reject = np.linspace(valor_t_critico, 4, 100)
plt.fill_between(x_reject, t.pdf(x_reject, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alpha})')

# Valor crítico
plt.axvline(valor_t_critico, color='red', linestyle='--', label=f'Valor Crítico t = {valor_t_critico:.4f}')

# Estadístico calculado
plt.axvline(estadistico_t_calculado, color='blue', linestyle='-', label=f'Estadístico t Calculado = {estadistico_t_calculado:.4f}')

# Marcar el valor p en la gráfica
x_p = np.linspace(estadistico_t_calculado, 4, 100)
plt.fill_between(x_p, t.pdf(x_p, grados_libertad), color='green', alpha=0.5, label=f'Valor p = {valor_p:.4f}')


plt.title('Prueba de Hipótesis Unilateral a la Derecha (t)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if estadistico_t_calculado > valor_t_critico:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que el tiempo promedio poblacional de entrega a domicilio es mayor que {media_poblacional_hipotetizada} minutos. Esto respalda la sospecha del cliente."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que el tiempo promedio poblacional de entrega a domicilio es mayor que {media_poblacional_hipotetizada} minutos. No hay suficiente evidencia para respaldar la sospecha del cliente de que el tiempo es mayor."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una media muestral de {media_muestra} minutos (o una más extrema) si la hipótesis nula (μ = {media_poblacional_hipotetizada} minutos) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia_alpha else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha}), {'rechazamos' if valor_p < nivel_significancia_alpha else 'no rechazamos'} la hipótesis nula.")
```

31. Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de
σ
2 = 2,5 minutos2
. Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral
de s
2 = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?


```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Datos del problema 31
varianza_poblacional_hipotetizada = 2.5  # minutos^2 (afirmación de la compañía)
varianza_muestra = 4  # minutos^2 (varianza de la muestra)
tamano_muestra = 16
nivel_significancia_alpha = 0.05

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La varianza poblacional es igual a 2.5 minutos^2.
# Hipótesis Alternativa (H1): La varianza poblacional es diferente a 2.5 minutos^2.
# Esta es una prueba bilateral (de dos colas).

print("1. Hipótesis:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} minutos²")
print(f"   H1: σ² ≠ {varianza_poblacional_hipotetizada} minutos² (Prueba bilateral)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Para probar la varianza poblacional con una muestra, usamos el estadístico Chi-cuadrado (χ²).
print("\n3. Estadístico de prueba: χ²")
print("   Distribución: Chi-cuadrado (χ²)")

# Grados de libertad
grados_libertad = tamano_muestra - 1
print(f"   Grados de libertad: {grados_libertad}")

# 4. Valores críticos y región de rechazo
# Para una prueba bilateral con α = 0.05 y grados de libertad = 15,
# el nivel de significancia se divide en dos colas (α/2 = 0.025).
# Buscamos los valores χ² tales que el área a la izquierda de χ²_critico_izq sea 0.025 y el área a la derecha de χ²_critico_der sea 0.025.
valor_chi2_critico_izq = chi2.ppf(nivel_significancia_alpha/2, grados_libertad)
valor_chi2_critico_der = chi2.ppf(1 - nivel_significancia_alpha/2, grados_libertad)

print(f"\n4. Valores críticos:")
print(f"   χ²_izq ({nivel_significancia_alpha/2:.3f}, {grados_libertad} gl) = {valor_chi2_critico_izq:.4f}")
print(f"   χ²_der ({1 - nivel_significancia_alpha/2:.3f}, {grados_libertad} gl) = {valor_chi2_critico_der:.4f}")
print(f"   Región de rechazo: χ²_calculado < {valor_chi2_critico_izq:.4f} o χ²_calculado > {valor_chi2_critico_der:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
estadistico_chi2_calculado = ((tamano_muestra - 1) * varianza_muestra) / varianza_poblacional_hipotetizada

# Para una prueba bilateral, el valor p es el doble del área en la cola correspondiente al estadístico χ² calculado.
# Si χ²_calculado es menor que la media (grados de libertad), calculamos el área a la izquierda y la duplicamos.
# Si χ²_calculado es mayor que la media, calculamos el área a la derecha y la duplicamos.
if estadistico_chi2_calculado < grados_libertad:
    valor_p = 2 * chi2.cdf(estadistico_chi2_calculado, grados_libertad)
else:
    valor_p = 2 * (1 - chi2.cdf(estadistico_chi2_calculado, grados_libertad))


print(f"\n5. Estadístico de prueba calculado: χ² = {estadistico_chi2_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000) # Ajustar rango x para la distribución chi2
plt.plot(x, chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} gl)')

# Región de rechazo (área en las colas)
x_reject_left = np.linspace(0, valor_chi2_critico_izq, 100)
plt.fill_between(x_reject_left, chi2.pdf(x_reject_left, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {nivel_significancia_alpha/2})')

x_reject_right = np.linspace(valor_chi2_critico_der, chi2.ppf(0.999, grados_libertad), 100)
plt.fill_between(x_reject_right, chi2.pdf(x_reject_right, grados_libertad), color='red', alpha=0.5) # No repetir label

# Valores críticos
plt.axvline(valor_chi2_critico_izq, color='red', linestyle='--', label=f'Valores Críticos χ² = {valor_chi2_critico_izq:.4f}, {valor_chi2_critico_der:.4f}')
plt.axvline(valor_chi2_critico_der, color='red', linestyle='--') # No repetir label

# Estadístico calculado
plt.axvline(estadistico_chi2_calculado, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2_calculado:.4f}')

# Marcar el valor p en la gráfica (área en las colas más extremas que el estadístico calculado)
if estadistico_chi2_calculado < grados_libertad: # Si el estadístico está a la izquierda de la media
     x_p_left = np.linspace(0, estadistico_chi2_calculado, 100)
     plt.fill_between(x_p_left, chi2.pdf(x_p_left, grados_libertad), color='green', alpha=0.5, label=f'Valor p/2 = {valor_p/2:.4f}')
else: # Si el estadístico está a la derecha de la media
    x_p_right = np.linspace(estadistico_chi2_calculado, chi2.ppf(0.999, grados_libertad), 100)
    plt.fill_between(x_p_right, chi2.pdf(x_p_right, grados_libertad), color='green', alpha=0.5, label=f'Valor p/2 = {valor_p/2:.4f}')


plt.title('Prueba de Hipótesis Bilateral (χ² para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if estadistico_chi2_calculado < valor_chi2_critico_izq or estadistico_chi2_calculado > valor_chi2_critico_der:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {varianza_poblacional_hipotetizada} minutos². Esto sugiere que la variabilidad en la duración de las llamadas difiere de lo afirmado por la compañía."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {varianza_poblacional_hipotetizada} minutos². No podemos afirmar que la variabilidad en la duración de las llamadas difiera significativamente de lo afirmado."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una varianza muestral de {varianza_muestra} minutos² (o una más extrema en cualquier dirección) si la hipótesis nula (σ² = {varianza_poblacional_hipotetizada} minutos²) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia_alpha else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha}), {'rechazamos' if valor_p < nivel_significancia_alpha else 'no rechazamos'} la hipótesis nula.")
```

32. Una máquina embotelladora debe mantener la varianza del llenado en σ
2 = 1,2 ml2
. Se toma una
muestra de n = 20 botellas y se encuentra s
2 = 2,0. ¿Existe evidencia al 1 % de que la varianza es
mayor?


```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Datos del problema 32
varianza_poblacional_hipotetizada = 1.2  # ml^2 (afirmación de la máquina)
varianza_muestra = 2.0  # ml^2 (varianza de la muestra)
tamano_muestra = 20
nivel_significancia_alpha = 0.01

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La varianza poblacional es igual a 1.2 ml^2.
# Hipótesis Alternativa (H1): La varianza poblacional es mayor que 1.2 ml^2.
# Esta es una prueba unilateral a la derecha.

print("1. Hipótesis:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} ml²")
print(f"   H1: σ² > {varianza_poblacional_hipotetizada} ml² (Prueba unilateral a la derecha)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Para probar la varianza poblacional con una muestra, usamos el estadístico Chi-cuadrado (χ²).
print("\n3. Estadístico de prueba: χ²")
print("   Distribución: Chi-cuadrado (χ²)")

# Grados de libertad
grados_libertad = tamano_muestra - 1
print(f"   Grados de libertad: {grados_libertad}")

# 4. Valor crítico y región de rechazo
# Para una prueba unilateral a la derecha con α = 0.01 y grados de libertad = 19,
# buscamos el valor χ² tal que el área a su derecha sea 0.01.
valor_chi2_critico = chi2.ppf(1 - nivel_significancia_alpha, grados_libertad)

print(f"\n4. Valor crítico: χ²_{1 - nivel_significancia_alpha:.3f} ({grados_libertad} gl) = {valor_chi2_critico:.4f}")
print(f"   Región de rechazo: χ²_calculado > {valor_chi2_critico:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
estadistico_chi2_calculado = ((tamano_muestra - 1) * varianza_muestra) / varianza_poblacional_hipotetizada

# El valor p para una prueba unilateral a la derecha es el área a la derecha del estadístico χ² calculado.
valor_p = 1 - chi2.cdf(estadistico_chi2_calculado, grados_libertad)

print(f"\n5. Estadístico de prueba calculado: χ² = {estadistico_chi2_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000) # Ajustar rango x para la distribución chi2
plt.plot(x, chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} gl)')

# Región de rechazo (área a la derecha del valor crítico)
x_reject = np.linspace(valor_chi2_critico, chi2.ppf(0.999, grados_libertad), 100)
plt.fill_between(x_reject, chi2.pdf(x_reject, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alpha})')

# Valor crítico
plt.axvline(valor_chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_chi2_critico:.4f}')

# Estadístico calculado
plt.axvline(estadistico_chi2_calculado, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2_calculado:.4f}')

# Marcar el valor p en la gráfica
x_p = np.linspace(estadistico_chi2_calculado, chi2.ppf(0.999, grados_libertad), 100)
plt.fill_between(x_p, chi2.pdf(x_p, grados_libertad), color='green', alpha=0.5, label=f'Valor p = {valor_p:.4f}')


plt.title('Prueba de Hipótesis Unilateral a la Derecha (χ² para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if estadistico_chi2_calculado > valor_chi2_critico:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional del llenado de botellas es mayor que {varianza_poblacional_hipotetizada} ml². Esto sugiere que la variabilidad en el llenado de las botellas es mayor de lo permitido."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional del llenado de botellas es mayor que {varianza_poblacional_hipotetizada} ml². No hay suficiente evidencia para afirmar que la variabilidad en el llenado de las botellas es significativamente mayor."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una varianza muestral de {varianza_muestra} ml² (o una más extrema) si la hipótesis nula (σ² = {varianza_poblacional_hipotetizada} ml²) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia_alpha else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha}), {'rechazamos' if valor_p < nivel_significancia_alpha else 'no rechazamos'} la hipótesis nula.")
```

33. El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de
al menos σ
2 = 0,5 mm2
. Una muestra de n = 12 piezas dio una varianza s
2 = 0,3. ¿Se puede
concluir al 5 % que la varianza poblacional es menor que la garantizada?


```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Datos del problema 33
varianza_poblacional_hipotetizada = 0.5  # mm^2 (afirmación del control de calidad - al menos)
varianza_muestra = 0.3  # mm^2 (varianza de la muestra)
tamano_muestra = 12
nivel_significancia_alpha = 0.05

# 1. Definición de hipótesis
# Hipótesis Nula (H0): La varianza poblacional es igual a 0.5 mm^2. (Asumimos el límite inferior de la afirmación)
# Hipótesis Alternativa (H1): La varianza poblacional es menor que 0.5 mm^2.
# Esta es una prueba unilateral a la izquierda.

print("1. Hipótesis:")
print(f"   H0: σ² = {varianza_poblacional_hipotetizada} mm²")
print(f"   H1: σ² < {varianza_poblacional_hipotetizada} mm² (Prueba unilateral a la izquierda)")

# 2. Nivel de significancia α
print(f"\n2. Nivel de significancia α: {nivel_significancia_alpha}")

# 3. Estadístico de prueba y distribución
# Para probar la varianza poblacional con una muestra, usamos el estadístico Chi-cuadrado (χ²).
print("\n3. Estadístico de prueba: χ²")
print("   Distribución: Chi-cuadrado (χ²)")

# Grados de libertad
grados_libertad = tamano_muestra - 1
print(f"   Grados de libertad: {grados_libertad}")

# 4. Valor crítico y región de rechazo
# Para una prueba unilateral a la izquierda con α = 0.05 y grados de libertad = 11,
# buscamos el valor χ² tal que el área a su izquierda sea 0.05.
valor_chi2_critico = chi2.ppf(nivel_significancia_alpha, grados_libertad)

print(f"\n4. Valor crítico: χ²_{nivel_significancia_alpha:.3f} ({grados_libertad} gl) = {valor_chi2_critico:.4f}")
print(f"   Región de rechazo: χ²_calculado < {valor_chi2_critico:.4f}")

# 5. Cálculo del estadístico de prueba y valor p
# Fórmula del estadístico Chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
estadistico_chi2_calculado = ((tamano_muestra - 1) * varianza_muestra) / varianza_poblacional_hipotetizada

# El valor p para una prueba unilateral a la izquierda es el área a la izquierda del estadístico χ² calculado.
valor_p = chi2.cdf(estadistico_chi2_calculado, grados_libertad)

print(f"\n5. Estadístico de prueba calculado: χ² = {estadistico_chi2_calculado:.4f}")
print(f"   Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado
plt.figure(figsize=(10, 6))
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000) # Ajustar rango x para la distribución chi2
plt.plot(x, chi2.pdf(x, grados_libertad), label=f'Distribución χ² ({grados_libertad} gl)')

# Región de rechazo (área a la izquierda del valor crítico)
x_reject = np.linspace(0, valor_chi2_critico, 100)
plt.fill_between(x_reject, chi2.pdf(x_reject, grados_libertad), color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alpha})')

# Valor crítico
plt.axvline(valor_chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_chi2_critico:.4f}')

# Estadístico calculado
plt.axvline(estadistico_chi2_calculado, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2_calculado:.4f}')

# Marcar el valor p en la gráfica
x_p = np.linspace(0, estadistico_chi2_calculado, 100)
plt.fill_between(x_p, chi2.pdf(x_p, grados_libertad), color='green', alpha=0.5, label=f'Valor p = {valor_p:.4f}')

plt.title('Prueba de Hipótesis Unilateral a la Izquierda (χ² para Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Decisión estadística y conclusión
print("\n7. Decisión estadística y Conclusión:")
if estadistico_chi2_calculado < valor_chi2_critico:  # Equivalente a valor_p < alpha
    decision = "Rechazar H0"
    conclusion = f"Hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional en el grosor de las piezas metálicas es menor que {varianza_poblacional_hipotetizada} mm². Esto sugiere que la variabilidad en el grosor es menor que la garantizada por el control de calidad."
else:
    decision = "No rechazar H0"
    conclusion = f"No hay suficiente evidencia estadística, con un nivel de significancia del {nivel_significancia_alpha*100}%, para concluir que la varianza poblacional en el grosor de las piezas metálicas es menor que {varianza_poblacional_hipotetizada} mm². No podemos afirmar que la variabilidad sea significativamente menor que la garantizada."

print(f"   Decisión: {decision}")
print(f"   Conclusión: {conclusion}")
print(f"\n   Significado del valor p ({valor_p:.4f}): El valor p es la probabilidad de obtener una varianza muestral de {varianza_muestra} mm² (o una más extrema) si la hipótesis nula (σ² = {varianza_poblacional_hipotetizada} mm²) fuera verdadera. Dado que el valor p ({valor_p:.4f}) es {'menor' if valor_p < nivel_significancia_alpha else 'mayor'} que el nivel de significancia (α = {nivel_significancia_alpha}), {'rechazamos' if valor_p < nivel_significancia_alpha else 'no rechazamos'} la hipótesis nula.")
```
